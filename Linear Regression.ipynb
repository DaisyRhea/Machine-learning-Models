```python
numeric=data.select_dtypes(include=('int64','float64'))
```

```python
categoric=data.select_dtypes(exclude=('int64','float64'))
```

```python
categoric.head()
```

```python
from sklearn import preprocessing as pre
```

```python
le=pre.LabelEncoder()
```

```python
categoric['Make']=le.fit_transform(categoric['Make'])
```

```python
for x in categoric.columns:
    categoric[x]=le.fit_transform(categoric[x])
```

```python
categoric.head()
```

```python
final_data=pd.concat([categoric,numeric],axis=1)
```

```python
final_data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
final_data.iloc[:,0:14]
```

```python
x_train,x_test,y_train,y_test=train_test_split(final_data.iloc[:,0:14],final_data['MSRP'],test_size=0.3,random_state=1)
```

```python
x_train.head()
```

```python
from sklearn import linear_model
from sklearn.metrics import r2_score
```

```python
clf=linear_model.LinearRegression()
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
print(r2_score(y_test,y_pred))
```
