```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("train_u6lujuX_CVtuZ9i (2).csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data['Gender'].fillna(data['Gender'].mode()[0],inplace=True)
```

```python
data['Married'].fillna(data['Married'].mode()[0],inplace=True)
```

```python
data['Dependents'].fillna(data['Dependents'].mode()[0],inplace=True)
```

```python
data['Self_Employed'].fillna(data['Self_Employed'].mode()[0],inplace=True)
```

```python
data['LoanAmount'].fillna(round(data['LoanAmount'].mean(),0),inplace=True)
```

```python
data['Loan_Amount_Term'].fillna(data['Loan_Amount_Term'].mode()[0],inplace=True)
```

```python
data['Credit_History'].fillna(data['Credit_History'].mode()[0],inplace=True)
```

```python
data.isna().sum()
```

```python
num_data=data.select_dtypes(include=('int64','float64'))
num_data
```

```python
cat_data=data.select_dtypes(exclude=('int64','float64'))
cat_data
```

```python
from sklearn import preprocessing as p
```

```python
le=p.LabelEncoder()
```

```python
for x in cat_data.columns:
    cat_data[x]=le.fit_transform(cat_data[x])
```

```python
cat_data
```

```python
data1=pd.concat([num_data,cat_data],axis=1)
```

```python
data1
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(data1.iloc[:,0:12],data1['Loan_Status'],test_size=0.2,random_state=100)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import confusion_matrix
logreg=LogisticRegression()
```

```python
logreg.score()
```

```python
B=logreg.fit(x_train,y_train)
```

```python
predicted=logreg.predict(x_test)
predicted
```

```python
print(metrics.confusion_matrix(y_test,predicted))
```

```python
print(logreg.score(x_test,y_test))
```

```python
print(logreg.score(x_train,y_train))
```

```python
print(metrics.accuracy_score(y_test,predicted))
```

```python
print(metrics.classification_report(y_test,predicted))
```
